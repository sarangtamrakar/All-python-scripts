```python
# change the format of of datetime
train["datetime"] = pd.to_datetime(train["datetime"])
```

```python
train["hour"] = train["datetime"].dt.hour
train["year"] = train.datetime.dt.year

df1 = train[train["year"] == 2011]
df2 = train[train["year"]==2012]
```

```python
_=sns.catplot("hour","count",data=df1,kind="bar",palette="rainbow",height=10,aspect=2)
plt.title("Avg count by  hour in 2011")
```

```python
_=sns.catplot("hour","count",data=df2,kind="bar",palette="rainbow",height=8,aspect=2)
plt.title("Avg count by hour in 2012")
```

```python
_=sns.catplot("hour","count",data=train,kind="bar",palette="winter",height=8,aspect=2,hue="year",)
plt.title("Avg count By hour in 2011 & 2012")
```

```python
_=sns.catplot("hour","count",data=df1,kind="box",palette="rainbow",height=8,aspect=2)
plt.title("count By hour in 2011")
```
