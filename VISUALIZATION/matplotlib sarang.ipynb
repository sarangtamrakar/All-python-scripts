```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
df
```

### matplotlib

```python
plt.style.available
```

```python
x=df['petal_width']
y1=df['sepal_length']
y2=df['sepal_width']
```

```python
# line plot it is bivariate analysis,,it means it take two argumnt for numerical
plt.figure(figsize=(10,5))
plt.style.use('seaborn')
plt.plot(x,y1,label='sepal length',alpha=1)
plt.plot(x,y2,label='sepal width',alpha=.8)
plt.xlabel('petal width')
plt.ylabel('axis y')
plt.title("it's my plot")
plt.xlim(0,2.8)
plt.ylim(0,9)
plt.legend()
plt.grid(True)
plt.show()
```

```python
#bar plot it is bivariate analysis for numerical data type
plt.style.use('ggplot')
plt.bar(x,y1,label='sepel length',edgecolor='black',alpha=1)
plt.bar(x,y2,label='sepal width',edgecolor='white')
plt.xlabel('x axis')
plt.ylabel('y label')
plt.title('my bar plot')
plt.tight_layout()
plt.show()
```

```python
#barh plot it is bivariate analysis for numerical data type
plt.style.use('ggplot')
plt.barh(x,y1,label='sepel length',edgecolor='black',alpha=1)
plt.barh(x,y2,label='sepal width',edgecolor='white')
plt.xlabel('x axis')
plt.ylabel('y label')
plt.title('my bar plot')
plt.tight_layout()
plt.show()
```

```python
# histogram plot for univariate analysis, it takes one argument & give the frequency

plt.hist(x,bins=5,edgecolor='black',label='petal width',orientation='vertical',stacked=True,log=True)
plt.hist(y2,bins=5,edgecolor='blue',label='sepal width',orientation='vertical',stacked=True,log=True)
plt.xlabel('x axis')
plt.ylabel('frequency')
plt.title('it is m histogram')
plt.axvline(x.mean())
plt.axvline(y2.mean())
plt.legend()
plt.show()
```

```python
# scatter plot is used for bivariate analysis
plt.figure(figsize=(10,5))
plt.style.use('fivethirtyeight')
plt.scatter(x,y1,s=50,c='blue',alpha=.75,label='sepal length',marker='*')
plt.scatter(x,y2,s=70,c='red',alpha=.75,label='sepal width',marker='^')
plt.xlabel('petal width')
plt.ylabel('y axis')
plt.title('it is my scatter plot')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# hexbin similar to the scatter plot
plt.hexbin(x,y2,gridsize=5,bins=10)
plt.title('it is my hexbin plot')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

```python
# stack plot for bivariate analysis
plt.figure(figsize=(13,5))
plt.stackplot(x,y1,y2,labels=['sepal length','sepal width'])
plt.title('it is my stack plot')
plt.xlabel('petal width')
plt.ylabel('y axis')
plt.legend(loc=('upper left'))
plt.tight_layout()
plt.show()
```
